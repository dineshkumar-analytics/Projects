```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

OLD_FILE = "Old_Files.xlsx"
NEW_FILE = "New_Files.xlsx"
OUTPUT = "teams_difference_highlight_2.xlsx"

# ---- Load Excel files ----
old_df = pd.read_excel(OLD_FILE)
new_df = pd.read_excel(NEW_FILE)

# ---- Align lengths ----
max_len = max(len(old_df), len(new_df))
old_df = old_df.reindex(range(max_len))
new_df = new_df.reindex(range(max_len))

# ---- Combine into one DataFrame ----
# Column order: TEAM A OLD | TEAM B OLD | TEAM A NEW | TEAM B NEW
out = pd.concat([old_df.add_suffix("_OLD"), new_df.add_suffix("_NEW")], axis=1)

# ---- Save WITHOUT headers ----
out.to_excel(OUTPUT, index=False, header=False)

# ---- Highlight differences ----
wb = load_workbook(OUTPUT)
ws = wb.active

yellow = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

for row in range(1, max_len + 1):
    # TEAM A
    v1 = ws[f"A{row}"].value
    v2 = ws[f"C{row}"].value
    if v1 != v2:
        ws[f"A{row}"].fill = yellow
        ws[f"C{row}"].fill = yellow

    # TEAM B
    v1 = ws[f"B{row}"].value
    v2 = ws[f"D{row}"].value
    if v1 != v2:
        ws[f"B{row}"].fill = yellow
        ws[f"D{row}"].fill = yellow

wb.save(OUTPUT)
print("\nðŸŽ¯ Done â€” differences highlighted in_1:")
print(OUTPUT)
```
